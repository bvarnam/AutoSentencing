```python
features = ['INOUT','INTDUM','METHMIN','MONRACE',
            'MONSEX','MWEIGHT','NEWCIT','NEWCNVTN','NEWEDUC',
            'NEWRACE','NODRUG','NUMDEPEN','OFFGUIDE', 'PRISDUM', 'PROBATN','PROBDUM',
            'QUARTER'
           ]

b_list = ['ACCCAT', 'ACCGDLN', 'AGE', 'AGECAT', 'ALTDUM', 'AMENDYR', 'AMTTOTAL', 'CASETYPE', 
          'CITWHERE', 'COMBDRG2', 'COMDUM', 'CRIMHIST', #'DEFCONSL', 
          'DISPOSIT', 'DISTRICT', 'DRUGMIN', 'DSPLEA', 'DUFROM1-DUFROMX', 'DUTO1-DUFOX', 
          'EDUCATN', 'HOMDUM', 'HRCOMSRV']

c_list = ['RANGEPT', 'REAS1-REASX', 'REGSXMIN', 'REL2PTS', 'RELMIN', 'RESTDET1-RESTDETX',
         'RESTDUM', 'SAFE', 'SAFETY', 'SENSPCAP', 'SENSPLT0', 
          #'SENTDATE', 
          'SENTIMP',
         #'SENTTOT0', 
          'SMAX1-SMAXX',  'SMIN1-SMINX', 'SOURCES', 'STATMAX', 'STATMIN',
         'SUPERMAX', 'SUPERMIN', 'SUPRDUM', 'SUPREL', 'TIMSERVC', 'TOTCHPTS', 'TOTREST',
         'TOTUNIT', 'TYPEMONY', 'TYPEOTHS', 'UNIT1-UNITX', 'WGT1-WGTX', 'XCRHISSR', 'XFOLSOR',
         'XMAXSOR', 'XMINSOR', 'SENTRNGE' ]
```

```python
feats = b_list + features + c_list
```

```python
def get_cols(col_list):
    cols = []
    for col in col_list:
        if '-' in col:
            col = col.split('-')[0][:-1]
            features = [feat for feat in df.columns if col in feat]
            for f in features:
                cols.append(f)
        else:
            cols.append(col)
    return cols
```

```python
df_pruned = df[get_cols(feats)]
```

```python
drugs = df_pruned[(df_pruned['OFFGUIDE'] == 9) | (df_pruned['OFFGUIDE'] == 10)]
```
